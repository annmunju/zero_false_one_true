```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

### 로지스틱 회귀와 KNN 기반 (소프트) 보팅

```python
def voting_lr_knn(X_train, X_test, y_train, y_test, voting='soft', print_all=True):
    lr_clf = LogisticRegression()
    knn_clf = KNeighborsClassifier()
    vo_clf = VotingClassifier(estimators = [('LR', lr_clf), ('KNN', knn_clf)], voting=voting)

    classifiers = [vo_clf, lr_clf, knn_clf]

    if print_all:
        clf_ls = []
        for classifier in classifiers:
            classifier.fit(X_train, y_train)
            pred = classifier.predict(X_test)
            print(classifier.__class__.__name__, '정확도 :', accuracy_score(y_test, pred))
            print(classifier.__class__.__name__, '정밀도 :', precision_score(y_test, pred))
            print(classifier.__class__.__name__, '재현율 :', recall_score(y_test, pred))
            print(classifier.__class__.__name__, '오차 행렬 :\n', confusion_matrix(y_test, pred))
            print()
    else:
        return vo_clf
```

```python
def Skfold_voting_lr_knn(X, y, n_splits=5, voting='soft'):

    st_kfold = StratifiedKFold(n_splits = n_splits)

    cv_acc, cv_pre, cv_recall = [], [], []

    n = 1
    for train_idx , val_idx in st_kfold.split(X , y):
        
        X_train = X.iloc[train_idx]
        X_test = X.iloc[val_idx]
        y_train = y.iloc[train_idx]
        y_test = y.iloc[val_idx]
        
        vo_clf = voting_lr_knn(X_train, X_test, y_train, y_test, voting=voting, print_all=False)
        vo_clf.fit(X_train , y_train)
        pred = vo_clf.predict(X_test)
        acc = np.round(accuracy_score(y_test , pred), 3)
        pre = np.round(precision_score(y_test, pred), 3)
        recall = np.round(recall_score(y_test, pred), 3)
        matrix = confusion_matrix(y_test, pred)
        
        print(f'[교차검증 : {n}회]')
        print(f'{n}회 교차검증의 정확도 : {acc}')
        print(f'{n}회 교차검증의 정밀도 : {pre}')
        print(f'{n}회 교차검증의 재현율 : {recall}')
        print(f'{n}회 교차검증의 오차행렬 : \n{matrix} \n')
        
        n+=1

        cv_acc.append(acc)
        cv_pre.append(pre)
        cv_recall.append(recall)


    print('평균 정확도' , np.round( np.mean(cv_acc),2))
    print('평균 정밀도' , np.round( np.mean(cv_pre),2))
    print('평균 재현율' , np.round( np.mean(cv_recall),2))
```

### 로지스틱 회귀와 의사결정 나무 기반 (소프트) 보팅

```python
def voting_lr_dt(X_train, X_test, y_train, y_test, voting='soft', print_all=True):
    lr_clf = LogisticRegression()
    dt_clf = DecisionTreeClassifier()
    vo_clf = VotingClassifier(estimators = [('LR', lr_clf), ('DT', dt_clf)], voting=voting)

    classifiers = [vo_clf, lr_clf, dt_clf]

    if print_all:
        for classifier in classifiers:
            classifier.fit(X_train, y_train)
            pred = classifier.predict(X_test)
            print(classifier.__class__.__name__, '정확도 :', accuracy_score(y_test, pred))
            print(classifier.__class__.__name__, '정밀도 :', precision_score(y_test, pred))
            print(classifier.__class__.__name__, '재현율 :', recall_score(y_test, pred))
            print(classifier.__class__.__name__, '오차 행렬 :\n', confusion_matrix(y_test, pred))
            print()
    else:
        return vo_clf
```

```python
def Skfold_voting_lr_dt(X, y, n_splits=5, voting='soft', test_size=0.3):

    st_kfold = StratifiedKFold(n_splits = n_splits)

    cv_acc, cv_pre, cv_recall = [], [], []

    n = 1
    for train_idx , val_idx in st_kfold.split(X , y):
        
        X_train = X.iloc[train_idx]
        X_test = X.iloc[val_idx]
        y_train = y.iloc[train_idx]
        y_test = y.iloc[val_idx]
        
        vo_clf = voting_lr_dt(X_train, X_test, y_train, y_test, voting=voting, print_all=False)
        vo_clf.fit(X_train , y_train)
        pred = vo_clf.predict(X_test)
        acc = np.round(accuracy_score(y_test , pred), 3)
        pre = np.round(precision_score(y_test, pred), 3)
        recall = np.round(recall_score(y_test, pred), 3)
        matrix = confusion_matrix(y_test, pred)
        
        print(f'[교차검증 : {n}회]')
        print(f'{n}회 교차검증의 정확도 : {acc}')
        print(f'{n}회 교차검증의 정밀도 : {pre}')
        print(f'{n}회 교차검증의 재현율 : {recall}')
        print(f'{n}회 교차검증의 오차행렬 : \n{matrix} \n')
        
        n+=1

        cv_acc.append(acc)
        cv_pre.append(pre)
        cv_recall.append(recall)


    print('평균 정확도' , np.round( np.mean(cv_acc),2))
    print('평균 정밀도' , np.round( np.mean(cv_pre),2))
    print('평균 재현율' , np.round( np.mean(cv_recall),2))
```
