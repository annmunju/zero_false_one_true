# 전처리 항목 정하기

1. 연령 소수점까지 있음(버림)

```python
df2016['연령'] = np.trunc(df2016['연령'])
df2017['연령'] = np.trunc(df2017['연령'])
df2018['연령'] = np.trunc(df2018['연령'])
```

2. 취업여부 1,2라고 응답한 사람 취업(1), 이외 미취업(0) : target값

```python
df2016.loc[(df2016['취업여부'] == 1) | (df2016['취업여부'] == 2),'취업여부'] = 1
df2017.loc[(df2017['취업여부'] == 1) | (df2017['취업여부'] == 2),'취업여부'] = 1
df2018.loc[(df2018['취업여부'] == 1) | (df2018['취업여부'] == 2),'취업여부'] = 1
```

```python
df2016.loc[(df2016['취업여부'] != 1), '취업여부'] = 0
df2017.loc[(df2017['취업여부'] != 1), '취업여부'] = 0
df2018.loc[(df2018['취업여부'] != 1), '취업여부'] = 0
```

```python
print(df2016['취업여부'].value_counts())
print(df2017['취업여부'].value_counts())
print(df2018['취업여부'].value_counts())
```

3. 응답하지 않은 경우와 모름의 경우 -1, 없음의 경우 -3 -> 0으로 수정. 같은 이유로 NaN값인 경우 0으로 수정

```python
na_To_zero = ['산업분류', '직업분류', '사업체위치(시도)', '종사상 지위', '계약여부', '정규직여부', '인턴경험여부(현직장)', '근로소득(현재_월_만원)',
       '근로소득(초임_월_만원)', '구직기간(개월)', '구직기간(주)', '구직경로', '구직방법', '신입여부',
       '취업_중요요소_1순위', '취업_중요요소_2순위', '첫직장여부']
```

```python
for col in na_To_zero:

    df2016.loc[df2016[col] <= -1, col] = 0
    df2016[col].fillna(0,inplace=True)

    df2017.loc[df2017[col] <= -1, col] = 0
    df2017[col].fillna(0,inplace=True)

    df2018.loc[df2018[col] <= -1, col] = 0
    df2018[col].fillna(0,inplace=True)
```

```python
na_cnt_df = pd.DataFrame()
na_cnt_df['2016'] = pd.DataFrame(df2016.isnull().sum())
na_cnt_df['2017'] = pd.DataFrame(df2017.isnull().sum())
na_cnt_df['2018'] = pd.DataFrame(df2018.isnull().sum())
na_cnt_df.iloc[11:-5]
```

4. 직원수(기업체), 직원수(사업체) 분리되어 있음.. 굳이? 

- 기업체만 남기기
- NaN값 0으로
- 직원수(기업체) -> 직원수 명칭 변경

```python
df2016.drop(columns='직원수(사업체)', inplace=True)
df2017.drop(columns='직원수(사업체)', inplace=True)
df2018.drop(columns='직원수(사업체)', inplace=True)
```

```python
df2016.loc[df2016['직원수(기업체)'] == -1, '직원수(기업체)'] = 0
df2016['직원수(기업체)'].fillna(0,inplace=True)

df2017.loc[df2017['직원수(기업체)'] == -1, '직원수(기업체)'] = 0
df2017['직원수(기업체)'].fillna(0,inplace=True)

df2018.loc[df2018['직원수(기업체)'] == -1, '직원수(기업체)'] = 0
df2018['직원수(기업체)'].fillna(0,inplace=True)
```

```python
df2016.rename({'직원수(기업체)':'직원수'}, axis=1, inplace=True)
df2017.rename({'직원수(기업체)':'직원수'}, axis=1, inplace=True)
df2018.rename({'직원수(기업체)':'직원수'}, axis=1, inplace=True)
```

5. 사업체 위치 국내외 구분에서 국외는 15~20명으로 유의미한 지표가 될 것 같지는 않음 (삭제)

```python
df2016.drop(columns='사업체위치(국내외)', inplace=True)
df2017.drop(columns='사업체위치(국내외)', inplace=True)
df2018.drop(columns='사업체위치(국내외)', inplace=True)
```

6. 사업체 위치 구시군은 너무 나눠져있음 분석에 어려울듯

```python
df2016.drop(columns='사업체위치(구시군)', inplace=True)
df2017.drop(columns='사업체위치(구시군)', inplace=True)
df2018.drop(columns='사업체위치(구시군)', inplace=True)
```

7. 구직기간 개월+주 단위를 합쳐 하나의 주단위 구직기간 컬럼 만들기

- 1개월 = 4.34주 로 계산


```python
df2016['구직기간(주)'] = round(df2016['구직기간(개월)'] * 4.34 + df2016['구직기간(주)'])
df2017['구직기간(주)'] = round(df2017['구직기간(개월)'] * 4.34 + df2017['구직기간(주)'])
df2018['구직기간(주)'] = round(df2018['구직기간(개월)'] * 4.34 + df2018['구직기간(주)'])
```

```python
df2017['구직기간(주)'].value_counts()
```

```python
df2016.drop(columns='구직기간(개월)', inplace=True)
df2017.drop(columns='구직기간(개월)', inplace=True)
df2018.drop(columns='구직기간(개월)', inplace=True)
```
